```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd

df['Highest_Ratio'] = 0

all_queries = list(df['Variation_Query']) + list(df1['Pre_Resolved_Query'])

tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform(all_queries)

# Split the TF-IDF matrix back into two parts: one for variation queries and one for resolved queries
tfidf_variation = tfidf_matrix[:len(df)]  
tfidf_resolved = tfidf_matrix[len(df):]   

# Calculate the cosine similarity between each variation query and each resolved query
for i in range(len(df)):
    for j in range(len(df1)):
        # Compute cosine similarity and convert to scalar from a 2D array
        cosine_sim = cosine_similarity(tfidf_variation[i], tfidf_resolved[j])[0][0]
        print(f"Cosine Similarity between '{df['Variation_Query'][i]}' and '{df1['Pre_Resolved_Query'][j]}': {cosine_sim}")

        if cosine_sim > df.loc[i, 'Highest_Ratio']:
            df.loc[i, 'Highest_Ratio'] = cosine_sim
            df.loc[i, 'Resolved_Index_tfidf'] = j + 1  

print(df[['Variation_Query', 'Resolved_Index_tfidf', 'Highest_Ratio']])

    
```
